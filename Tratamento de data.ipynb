Mesmo sendo uma base simples ela apreseta um problema serio que impossibilita uma análise 
Qual o problema?
O problema de é a coluna "date" que tem datas no formato BR e também no formarto USA
Vamos tratar?

```python
df = pd.read_csv(r"C:\Users\saull\Documents\Planilhas power bi\Sales\Balaji Fast Food Sales.csv")
df
```

```python
#Apagando coluna inutil
df = df.drop("order_id",axis=1)
```

```python
df.isnull().sum()
```

```python
#Apagando dados faltantes
df.dropna(inplace = True)
```

```python
df.head()
```

# Tratando o problema

```python

lista = []
for i in df['date']:
    if '-' in i:
        data = i.split('-')
        data[0], data[1] = data[1], data[0]
        data = '/'.join(data)
        # Condição para encontrar a linha desejada
        condicao = df['date'] == i
        # Mudando o valor da célula na coluna 'date' onde a condição é verdadeira
        df.loc[condicao, 'date'] = data

# Transformando coluna em Datetime
df['date'] = pd.to_datetime(df['date'])
        
```
